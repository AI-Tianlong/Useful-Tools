```python
import os
from PIL import Image
import numpy as np

original_labels_path = 'D:/ATL/SIRS/chusai/FuSai/Dataset/fusai_release/train/labels'

labels_list = os.listdir(original_labels_path)

labels_save_path = 'D:/ATL/SIRS/chusai/FuSai/Dataset/fusai_release/train/labels_rgb'


# 标签与类别序号对应
# 背景 000
# 水体 101
# 道路 202
# 建筑物 303
# 机场 204
# 火车站 205
# 光伏 806
# 停车场 807
# 操场 808
# 普通耕地 409
# 农业大棚 410
# 自然草地 511
# 绿地绿化 512
# 自然林 613
# 人工林 614
# 自然裸土 715
# 人为裸土 716
# 其他无法确定归属物 817

class_list = [0,101, 202, 303, 204, 205, 806, 807, 808,409, 410, 511, 512, 613, 614, 715, 716, 817]

print(len(class_list))

for i in range(len(labels_list)):

    label = np.array(Image.open(os.path.join(original_labels_path, labels_list[i])))

    for j in range(len(class_list)):

        label[np.where(label==class_list[j])] = j

    label = Image.fromarray(label)
    label.save(os.path.join(labels_save_path, labels_list[i]))

    print("已转化完成{}张标签！！！".format(i))




```
